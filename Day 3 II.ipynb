```python
import matplotlib.pyplot as plt
import pandas as pd
data = {
    "Year": [2000, 2005, 2010, 2015, 2020, 2025],
    "Average Income": [30000, 35000, 40000, 45000, 50000, 55000]
}
df = pd.DataFrame(data)
plt.figure(figsize=(10, 6))
plt.plot(df['Year'], df['Average Income'], marker='o', linestyle='-', color='green', label='Average Income')
plt.title('Change in Average Income Over the Years')
plt.xlabel('Year')
plt.ylabel('Average Income (USD)')
plt.grid(True)
plt.legend()
plt.show()
```
