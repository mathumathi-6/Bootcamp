```python
import matplotlib.pyplot as plt
import pandas as pd
data = {
    "Year": [2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022],
    "Finland": [7.4, 7.5, 7.6, 7.6, 7.8, 7.8, 7.9, 7.8],
    "Denmark": [7.5, 7.5, 7.5, 7.6, 7.6, 7.6, 7.6, 7.6],
    "Switzerland": [7.5, 7.5, 7.5, 7.5, 7.5, 7.5, 7.5, 7.5],
    "Iceland": [7.4, 7.5, 7.5, 7.5, 7.5, 7.5, 7.5, 7.5],
    "Netherlands": [7.4, 7.4, 7.4, 7.5, 7.5, 7.5, 7.4, 7.4]
}
df = pd.DataFrame(data)
plt.figure(figsize=(12, 8))
for country in df.columns[1:]:
    plt.plot(df['Year'], df[country], marker='o', linestyle='-', label=country)
plt.title('Happiness Score Trends (WHR) Over the Years')
plt.xlabel('Year')
plt.ylabel('Happiness Score (WHR)')
plt.grid(True)
plt.legend(title='Countries')
plt.show()
```
