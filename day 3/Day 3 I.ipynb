```python
import matplotlib.pyplot as plt
import pandas as pd
data = {
    "Year": [1950, 1960, 1970, 1980, 1990, 2000, 2010, 2020],
    "Life Expectancy": [48.0, 52.5, 58.0, 63.0, 65.5, 67.5, 70.0, 72.8]
}
df = pd.DataFrame(data)
plt.figure(figsize=(10, 6))
plt.plot(df['Year'], df['Life Expectancy'], marker='o', linestyle='-', color='blue', label='Life Expectancy')
plt.title('Change in Average Life Expectancy Over the Years')
plt.xlabel('Year')
plt.ylabel('Average Life Expectancy (Years)')
plt.grid(True)
plt.legend()
plt.show()
```
