```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
data = {
    "Country": ["USA", "India", "China", "Brazil", "Nigeria"],
    "Continent": ["North America", "Asia", "Asia", "South America", "Africa"],
    "Income_2010": [45000, 3000, 8000, 10000, 2000],
    "Income_2020": [60000, 5000, 12000, 15000, 2500],
    "Life_Expectancy_2010": [78, 65, 74, 72, 52],
    "Life_Expectancy_2020": [79, 69, 77, 75, 55],
    "Population_2010": [309000000, 1230000000, 1340000000, 195000000, 158000000],
    "Population_2020": [331000000, 1380000000, 1440000000, 213000000, 206000000]
}
df = pd.DataFrame(data)
selected_year = 2020  # Change this to 2010 or 2020 to simulate the dropdown
income_col = f"Income_{selected_year}"
life_exp_col = f"Life_Expectancy_{selected_year}"
pop_col = f"Population_{selected_year}"
bubble_size = np.sqrt(df[pop_col] / 1000000) * 100
continent_colors = {
    "North America": "blue",
    "Asia": "red",
    "South America": "green",
    "Africa": "orange"
}
colors = df["Continent"].map(continent_colors)
plt.figure(figsize=(12, 8))
plt.scatter(
    df[income_col],
    df[life_exp_col],
    s=bubble_size,
    c=colors,
    alpha=0.6,
    edgecolors="w",
    linewidth=0.5
)
for i in range(df.shape[0]):
    plt.text(
        df[income_col][i],
        df[life_exp_col][i],
        df["Country"][i],
        fontsize=9,
        ha="center"
    )
plt.xscale('log')
plt.title(f'Life Expectancy vs Income ({selected_year})')
plt.xlabel('Income per Capita (Log Scale)')
plt.ylabel('Life Expectancy (Years)')
plt.grid(True, linestyle='--', alpha=0.5)
for continent, color in continent_colors.items():
    plt.scatter([], [], c=color, alpha=0.6, s=100, label=continent)
plt.legend(title='Continent')
plt.show()
```
